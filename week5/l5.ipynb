```python
import torch
import torch.nn.functional as F
```

```python
x = torch.tensor([-2.0, -1.0, 0.0, 1.0, 2.0])
```

```python
sigmoid = torch.sigmoid(x)
tanh = torch.tanh(x)
relu = F.relu(x)
leaky_relu = F.leaky_relu(x, negative_slope=0.1)
softmax = F.softmax(x, dim=0)  # Softmax over the tensor elements
```

```python
print("Input:", x)
print("Sigmoid:", sigmoid)
print("Tanh:", tanh)
print("ReLU:", relu)
print("Leaky ReLU:", leaky_relu)
print("Softmax:", softmax)
```

```python
import matplotlib.pyplot as plt

# Data
epochs = [1, 2, 3, 4, 5]
loss = [0.6830, 0.6382, 0.6020, 0.5469, 0.5029]

# Plot
plt.plot(epochs, loss, marker='o', linestyle='-', color='blue')
plt.title('Epoch vs Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.grid(True)
plt.show()
```
